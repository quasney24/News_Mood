```python

# Dependencies
import tweepy
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import requests as req
analyzer = SentimentIntensityAnalyzer()
```

```python
# Twitter API Keys
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
target = ["@BBC", "@CBSNews", "@CNN", "@FoxNews", "@nytimes"]

sentiments = []
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
for user in target:

    counter = 1

    
    public_tweets = api.user_timeline(user, count=100, result_type="recent")
    for tweet in public_tweets:
        compound = analyzer.polarity_scores(tweet["text"])["compound"]
        pos = analyzer.polarity_scores(tweet["text"])["pos"]
        neu = analyzer.polarity_scores(tweet["text"])["neu"]
        neg = analyzer.polarity_scores(tweet["text"])["neg"]
        Counter2=counter

        sentiments.append({
            "User": (tweet['user']['name']),
           "Compound": compound,
           "Positive": pos,
           "Negative": neu,
           "Neutral": neg,
            "Counter2":counter})

        counter= counter+1
```

```python
sentiment_pd = pd.DataFrame.from_dict(sentiments)

sentiment_pd.head()
```

```python
BBC = sentiment_pd.loc[sentiment_pd['User'] == 'BBC']

FOX = sentiment_pd.loc[sentiment_pd['User'] == 'Fox News']

CBS = sentiment_pd.loc[sentiment_pd['User'] == 'CBS News']

CNN = sentiment_pd.loc[sentiment_pd['User'] == 'CNN']

NYT = sentiment_pd.loc[sentiment_pd['User'] == 'The New York Times']

sns.set(style="darkgrid")
plt.figure(figsize = (12,10))

plt.scatter(BBC["Counter2"], BBC["Compound"], label = "BBC", marker="o", s=200)
plt.scatter(FOX["Counter2"], FOX["Compound"], label = "Fox News", marker="o", s=200)
plt.scatter(CNN["Counter2"], CNN["Compound"], label = "CNN", marker="o", s=200)
plt.scatter(NYT["Counter2"], NYT["Compound"], label = "The New York Times", marker="o",s=200)
plt.scatter(CBS["Counter2"], CBS["Compound"], label = "CBS", marker="o", s=200)

plt.xlim(-5, counter + 5)
plt.ylim(-1.25, 1.25)

plt.title("Sentiment Analysis")
plt.ylabel("VADER Sentiment Analysis")
plt.xlabel("Past Tweets")

plt.legend(title="Tweet Handles", bbox_to_anchor=(1, 1), edgecolor="white")

# Save 
plt.savefig("Sentiment_Analysis.png")


# Show plot
plt.show()
```

```python
#average
av_sentiment = pd.DataFrame(sentiment_pd.groupby("User")["Compound"].mean())
plt.figure(figsize = (7,5))

x_axis = np.arange(len(av_sentiment["Compound"]))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["BBC", "CBS", "CNN", "FOX", "NYT"])

plt.bar(x_axis, av_sentiment["Compound"])
plt.axhline(0)
plt.grid(False)


plt.title("Average Twitter Sentiment by Handle")
plt.ylabel("VADER Tweet Sentiment")
          
# Save
plt.savefig("av_sentiment.png")

plt.show()
```
